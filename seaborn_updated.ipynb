```python
import pandas as pd
import numpy as np

import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips)
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker',style='time')
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker',palette='ch:r=-0.8, l=0.85',size='size')
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,sizes=(15,200),size='size')
```

```python
from numpy.random import randn
```

```python
df=pd.DataFrame(dict(time=np.arange(500),value=randn(500).cumsum()))
df.head()
```

```python
sns.relplot(x='time',y='value',kind='line',data=df)
```

```python
df=pd.DataFrame(randn(500,2).cumsum(axis=0),columns=['time','value'])
df.head()
```

```python
sns.relplot(x='time',y='value',data=df,kind='line',sort = True)
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri,kind='line',hue='event')
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri,hue='region',style='event',kind='line',ci=False)
```

```python
dots=sns.load_dataset('dots').query("align=='dots'")
dots.head()
```

```python
sns.relplot(x='time',y='firing_rate',data=dots,kind='line',hue='coherence',style='choice')
```

```python
penguins= sns.load_dataset('penguins')
penguins.head()
```

```python
sns.set_style('darkgrid')
sns.scatterplot(data=penguins, x='flipper_length_mm', y='body_mass_g',hue='island',palette="Dark2")
sns.set_context("notebook")
```

```python
sns.histplot(data=penguins,x="body_mass_g",hue="sex",multiple="stack")
```

```python
sns.regplot(data=penguins, x="flipper_length_mm",y="body_mass_g")
```

```python
sns.boxplot(data=penguins,x="species",y="body_mass_g",hue="sex")
```

```python
sns.countplot(data=penguins, x="species", hue="island")
```

```python
penguins.head()
```

```python
columns = ["bill_length_mm",	"bill_depth_mm",	"flipper_length_mm"	, "body_mass_g"	]
sns.heatmap(data=penguins[columns].corr(),annot=True,cmap="Blues",vmin=-0.2,linewidth = "2",linecolor="black")
```

```python
sns.pairplot(data=penguins,hue="sex",diag_kind="kde",palette = "Dark2")
```
