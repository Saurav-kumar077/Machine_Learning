```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[10,20,15,25]
plt.plot(x,y)
plt.show()
```

```python
x=['Mon','Tue','Wed','Thr','Fri']
y=[10,16,7,20,12]
plt.title('Bakery sales of this week')
plt.xlabel('Day of the week')
plt.ylabel('sale per day')


plt.plot(x,y,color='green',linestyle='dotted',linewidth='3',marker='o',label='Sales data of last month')
plt.legend()
plt.grid(color='blue',linestyle=':',linewidth=0.5)
plt.xticks(['Mon','Tue','Wed','Thr','Fri'],['D1','D2','D3','D4','D5'])
plt.show()
```

```python
product=['A','B','C','D']
sales = [1000,1200,800,1500]
plt.bar(product,sales,color='orange',label='2025 sale',width=0.3)
plt.title("sale of each product")
plt.xlabel("Product Name")
plt.ylabel("Sale value")
plt.legend()
plt.show()
```

```python
regions = ['North','South','East','West']
revenue = [3000,2000,1500,800]
plt.pie(revenue,labels=regions,colors=['gold','skyblue','aqua','lightgreen'],autopct='%1.1f%%')
plt.title('Revenue contribution by regions')
plt.show()
```

```python
scores = [85, 78, 92, 66, 88, 73, 95, 81, 90, 69]
plt.hist(scores,color='Skyblue',bins=5,edgecolor='black',alpha=0.6)
plt.xlabel('scores of students')
plt.ylabel('Number of student')
plt.title('score distrubution by student')
plt.show()
```

```python
hours_studies=([1,2,3,4,5])
marks = ([50,60,30,80,90])

plt.scatter(hours_studies,marks,color = 'green',label='Student data',marker='o')
plt.xlabel('Hours_studies')
plt.ylabel('Marks')
plt.title('Student score')
plt.legend()
plt.grid()
plt.show()
```

```python
#subplot
#1st Method
x=([1,2,3,4])
y=([10,15,20,25])
plt.subplot(1,2,1)#1st row , 2nd col,1st subplot
plt.plot(x,y)
plt.title('Line chart')

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title('Bar chart')
plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(1, 2,figsize=(10,5))  # 1 row, 2 columns
#2nd Method
axs[0].plot([1, 2, 3], [4, 5, 6])
axs[1].bar(['A', 'B', 'C'], [3, 6, 2])

axs[0].set_title('Line chart')
axs[1].set_title('Bar chart')
plt.tight_layout()
plt.savefig('comparison.pdf',dpi=300,bbox_inches='tight')
plt.show()
```
